```python
import cv2
path = './Videos/11.png'
image = cv2.imread(path)
print(image.shape)
x,y,_ = image.shape
```

```python
start_point = (0,int(0.6*x))
end_point = (y,int(0.6*x))
print(start_point)
print(end_point)
color = (0, 255, 0)
thickness = 9
image = cv2.line(image, start_point, end_point, color, thickness)
```

```python
start_point = (0,int(0.5*x))
end_point = (y,int(0.5*x))
print(start_point)
print(end_point)
color = (0, 255, 0)
thickness = 9
image = cv2.line(image, start_point, end_point, color, thickness)
```

```python
import matplotlib.pyplot as plt
plt.imshow(image)
```
